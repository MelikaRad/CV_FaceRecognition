```python
import numpy as np
import cv2
import matplotlib.pyplot as plt

def visualize(input, faces, thickness=2):
    if faces[1] is not None:
        for idx, face in enumerate(faces[1]):
            coords = face[:-1].astype(np.int32)
            cv2.rectangle(input, (coords[0], coords[1]), (coords[0]+coords[2], coords[1]+coords[3]), (0, 255, 0), thickness)
            cv2.circle(input, (coords[4], coords[5]), 2, (255, 0, 0), thickness)
            cv2.circle(input, (coords[6], coords[7]), 2, (0, 0, 255), thickness)
            cv2.circle(input, (coords[8], coords[9]), 2, (0, 255, 0), thickness)
            cv2.circle(input, (coords[10], coords[11]), 2, (255, 0, 255), thickness)
            cv2.circle(input, (coords[12], coords[13]), 2, (0, 255, 255), thickness)

def visualize_recognition(input, faces, thickness=2):
    if faces[1] is not None:
        for idx, face in enumerate(faces[1]):
            coords = face[:-1].astype(np.int32)
            
            if idx in index[0]:
                cv2.rectangle(input, (coords[0], coords[1]), (coords[0]+coords[2], coords[1]+coords[3]), (0, 255, 0), thickness)
            else:
                cv2.rectangle(input, (coords[0], coords[1]), (coords[0]+coords[2], coords[1]+coords[3]), (0, 0, 255), thickness)

# loading the model
detector = cv2.FaceDetectorYN.create(
    "face_detection_yunet_2023mar.onnx",
    "",
    (320, 320),
    0.8,
    0.3,
    5000
)

# image 1 face detection
image1 = cv2.imread("images/chester.png")
img1 = image1.copy()
img1Width = int(img1.shape[1])
img1Height = int(img1.shape[0])
img1 = cv2.resize(img1, (img1Width, img1Height))

detector.setInputSize((img1Width, img1Height))
faces1 = detector.detect(img1)
assert faces1[1] is not None, 'Cannot find a face in image1'

visualize_detection(img1, faces1)

# image 2 face detection
image2 = cv2.imread("images/Linkin-Park.jpg")
img2 = image2.copy()

detector.setInputSize((img2.shape[1], img2.shape[0]))
faces2 = detector.detect(img2)
assert faces2[1] is not None, 'Cannot find a face in {}'.format(args.image2)

visualize_detection(img2, faces2)

# cropping faces of both images
recognizer = cv2.FaceRecognizerSF.create(
"face_recognition_sface_2021dec.onnx","")

face1_align = recognizer.alignCrop(image1, faces1[1][0])

face2_align = []
for i in range(len(faces2[1])):
    face2_align.append(recognizer.alignCrop(image2, faces2[1][i]))
face2_align = np.array(face2_align)

# Extract features of faces in both images
face1_feature = recognizer.feature(face1_align)

face2_feature = []
for i in range(len(faces2[1])):
    face2_feature.append(recognizer.feature(face2_align[i]))
face2_feature = np.array(face2_feature)

# comparing
l2_similarity_threshold = 1.128

l2_score = []
for i in range(len(faces2[1])):
    l2_score.append(recognizer.match(face1_feature, face2_feature[i], cv2.FaceRecognizerSF_FR_NORM_L2))
l2_score = np.array(l2_score)

idx = np.where(l2_score <= l2_similarity_threshold)

# the result
print('there exists {} face(s) recognized as the same person. recognized at index/indices {} '.format(len(idx[0]), idx[0]))

# showing the result
result_image = image2.copy()
visualize_recognition(result_image, faces2)

plt.figure(figsize=(10,15))
plt.subplot(311);plt.imshow(img1[:,:,::-1]); plt.title('face to be recognized');
plt.subplot(312);plt.imshow(img2[:,:,::-1]); plt.title('image with faces');
plt.subplot(313);plt.imshow(result_image[:,:,::-1]); plt.title('result of recognition');
```

### Real-time face recognition

on a web-cam:

```python
import numpy as np
import cv2
import matplotlib.pyplot as plt

image = cv2.imread('images/chester.png') # set your own photo :)
img1 = image.copy()
img1Width = int(img1.shape[1])
img1Height = int(img1.shape[0])
img1 = cv2.resize(img1, (img1Width, img1Height))

detector.setInputSize((img1Width, img1Height))
faces1 = detector.detect(img1)

cv2.namedWindow('face recognition')

if faces1[1] is not None: 

    # recognizer tool 
    recognizer = cv2.FaceRecognizerSF.create(
    "face_recognition_sface_2021dec.onnx","")

    # cropping the face in the image
    face1_align = recognizer.alignCrop(image, faces1[1][0])

    #---------------------------
    
    
    
    cap = cv2.VideoCapture(0)

    while True:
        ret, frame = cap.read()
        if not ret:
            break
            
        frame = cv2.flip(frame, 1)

        # frame face detection

        detector.setInputSize((frame.shape[1], frame.shape[0]))
        faces2 = detector.detect(frame)
        
        if faces2[1] is not None:

            # cropping faces of the frame

            face2_align = []
            for i in range(len(faces2[1])):
                face2_align.append(recognizer.alignCrop(frame, faces2[1][i]))
            face2_align = np.array(face2_align)

            # Extract features of faces in both images
            face1_feature = recognizer.feature(face1_align)

            face2_feature = []
            for i in range(len(faces2[1])):
                face2_feature.append(recognizer.feature(face2_align[i]))
            face2_feature = np.array(face2_feature)

            # comparing
            l2_similarity_threshold = 1.128

            l2_score = []
            for i in range(len(faces2[1])):
                l2_score.append(recognizer.match(face1_feature, face2_feature[i], cv2.FaceRecognizerSF_FR_NORM_L2))
            l2_score = np.array(l2_score)

            index = np.where(l2_score <= l2_similarity_threshold)


            # visualizing the result
            result_frame = frame.copy()
            visualize_recognition(result_frame, faces2)
            cv2.imshow('face recognition',result_frame)
            
        else:
            cv2.imshow('face recognition',frame)


        key = cv2.waitKey(1)
        if key == ord("q"):
            break

    cap.release()
    cv2.destroyAllWindows()
```
